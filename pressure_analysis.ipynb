```python
from statsbombpy import sb
import pandas as pd
```

```python
competitions = sb.competitions()
competitions.head()
```

```python
matches = sb.matches(competition_id=43, season_id= 106)
matches.head()
```

```python
final_match = matches[(matches.home_team == "Argentina" ) & (matches.away_team == "France")]
final_match
match_id = final_match.iloc[0]["match_id"]
```

```python
events = sb.events(match_id = match_id)
events.head(1000)
```

```python
events.columns
```

```python
events[['minute', 'second', 'team', 'player', 'type']].head(100)
```

```python
events['type'].value_counts()
```

```python
passes = events[events['type'] == "Pass"]
passes.head()
```

```python
events['time_seconds'] = events['minute'] * 60 + events['second']
events = events.sort_values("time_seconds")
events[['time_seconds','team','type']].head(20)
```

```python
shots = events[events['type'] == 'Shot']
shots.head()
```

```python
goal = shots[(shots['shot_outcome'] == 'Goal') & (shots['period'] != 5)]
goal[['minute','team','player']]
```

```python
events['is_goal'] = ((events['type'] == 'Shot') & (events['shot_outcome'] == 'Goal') & (events['period'] != 5))
events['arg_goal'] = ((events['is_goal']) & (events['team'] == 'Argentina')).cumsum()
events['fra_goal'] = ((events['is_goal']) & (events['team'] == 'France')).cumsum()
events[['minute','team','arg_goal','fra_goal']].tail(50)
```

```python
#events['score_diff'] = events['arg_goal'] - events['fra_goal']
events['time_norm'] = events['time_seconds']/ events['time_seconds'].max()
#events['pressure_v0'] = abs(events['score_diff']) * events['time_norm']
```

```python
#events['team_is_losing'] = (events['team'] == 'Argentina') & (events['score_diff'] < 0) | (events['team'] == 'France') & (events['score_diff'] > 0)
#events['is_draw'] = events['score_diff'] == 0
```

```python
#events['pressure_v1'] = ((events['team_is_losing']) * (1 + abs(events['score_diff'])) * (events['time_norm']))
```

```python
#events['pressure_v2'] = (1 + abs(events['score_diff'])) * events['time_norm']
```

```python
events['arg_score_diff'] = events['arg_goal'] - events['fra_goal']
events['fra_score_diff'] = events['fra_goal'] - events['arg_goal']
events['pressure_arg'] = (1 + (-events['arg_score_diff']).clip(lower=0)) * events['time_norm']
events['pressure_fra'] = (1 + (-events['fra_score_diff']).clip(lower=0)) * events['time_norm']
```

```python
events['pressure_team'] = 0
events.loc[events['team'] == 'Argentina', 'pressure_team'] = events['pressure_arg']
events.loc[events['team'] == 'France', 'pressure_team'] = events['pressure_fra']
```

```python
events[(events['minute'] >= 76) & (events['minute'] <= 78)][['minute','team','pressure_team']]
```

```python
key_events = events[events['type'].isin(['Pass', 'Shot', 'Carry'])]
key_events.head(5)
```

```python
key_events = key_events[['minute', 'team', 'type', 'pressure_team']]
key_events.head(20)
```

```python
avg_pressure = key_events.groupby(['type'])['pressure_team'].mean()
avg_pressure
```

```python
key_events['pressure_level_v0'] = 'Low'
key_events.loc[(key_events['pressure_team'] > 0.3) & (key_events['pressure_team'] <= 0.6), 'pressure_level_v0'] = 'Medium'
key_events.loc[key_events['pressure_team'] > 0.6, 'pressure_level_v0'] = 'High'
```

```python
key_events[['team','minute','pressure_team','pressure_level_v0']].head(800)
```

```python
count_actions_v0 = key_events.groupby(['pressure_level_v0','type']).size()
count_actions_v0
```

```python
percentage_v0 = count_actions_v0.groupby(level=0).apply(lambda x: x / x.sum())
percentage_v0
```

```python
percentage_v0 = percentage_v0.reset_index(level=0, drop=True)
percentage_v0 = percentage_v0.to_frame(name='percentage').reset_index()
percentage_v0['percentage'] = percentage_v0['percentage'] * 100
percentage_v0
```

```python
events['pressure_team'].describe()
```

```python
low_threshold = key_events['pressure_team'].quantile(0.25)
high_threshold = key_events['pressure_team'].quantile(0.75)
key_events['pressure_level_v1'] = 'Low'
key_events.loc[(key_events['pressure_team'] > low_threshold) & (key_events['pressure_team'] <= high_threshold), 'pressure_level_v1'] = 'Medium'
key_events.loc[key_events['pressure_team'] > high_threshold, 'pressure_level_v1'] = 'High'
key_events[['team','minute','pressure_team','pressure_level_v1']].head(800)
```

```python
count_actions_v1 = key_events.groupby(['pressure_level_v1','type']).size()
count_actions_v1
```

```python
percentage_v1 = count_actions_v1.groupby(level=0).apply(lambda x: x / x.sum())
percentage_v1
```

```python
percentage_v1 = percentage_v1.reset_index(level=0, drop=True)
percentage_v1 = percentage_v1.to_frame(name='percentage').reset_index()
percentage_v1['percentage'] = percentage_v1['percentage'] * 100
percentage_v1
```

```python
high_pressure = key_events[key_events['pressure_level_v1'] == 'High']
high_pressure.head()
```

```python
team_counts = high_pressure.groupby(['team', 'type']).size()
team_counts
```

```python
pressure_time = events.groupby(['minute', 'team'])['pressure_team'].mean().reset_index()
pressure_time.head()
```

```python
arg_pressure_time = pressure_time[pressure_time['team'] == 'Argentina'] 
fra_pressure_time = pressure_time[pressure_time['team'] == 'France']
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(arg_pressure_time['minute'], arg_pressure_time['pressure_team'], label='Argentina')
plt.plot(fra_pressure_time['minute'], fra_pressure_time['pressure_team'], label='France')
plt.xlabel('Minute')
plt.ylabel('Average Pressure')
plt.title('Pressure Over Time')
plt.legend()
plt.show()
```
